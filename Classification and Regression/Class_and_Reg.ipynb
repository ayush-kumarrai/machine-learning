```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data for classification and regression
# Classification data
class_data = np.random.randn(100, 2)  # 100 samples, 2 features
class_labels = np.random.randint(0, 2, 100)  # Binary labels (0 or 1)

# Regression data
reg_data = np.random.randn(100, 1)  # 100 samples, 1 feature
reg_targets = 2 * reg_data.squeeze() + np.random.randn(100)  # Linear relationship with noise

# Plotting classification data
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(class_data[class_labels == 0, 0], class_data[class_labels == 0, 1], color='blue', label='Class 0')
plt.scatter(class_data[class_labels == 1, 0], class_data[class_labels == 1, 1], color='red', label='Class 1')
plt.title('Classification Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()

# Plotting regression data
plt.subplot(1, 2, 2)
plt.scatter(reg_data, reg_targets, color='green', label='Regression Data')
plt.plot(reg_data, 2 * reg_data.squeeze(), color='black', linestyle='--', label='True Relationship')
plt.title('Regression Data')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.legend()

plt.tight_layout()
plt.show()
```
